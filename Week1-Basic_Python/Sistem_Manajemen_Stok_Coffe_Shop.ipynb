## Sistem Manajemen Stok Coffee Shop

```python
jenis_kopi = []

def menu():
    print("\n--- Sistem Manajemen Stok Coffee Shop ---")
    print("-----------------------------------------")
    print("1. Lihat stok")
    print("2. Tambah stok")
    print("3. Cek ketersediaan")
    print("4. Hitung Kebutuhan")
    print("5. Keluar")

def menu_tambah():
    print("\nMenu Tambah Stok")
    print("a. tambah stok")
    print("b. tambah jenis kopi & stok\n")

def lihat_stok():
    if jenis_kopi:
        print("\n--- Daftar Biji Kopi ---")
        for nama, stok in jenis_kopi:
            print(f"{nama} : {stok}")
    else:
        print("Daftar jenis kopi kosong")

def tambah_stok():
    menu_tambah()
    pilih_tambah = input("Masukkan pilihan anda [a, b]: ")
    if pilih_tambah.lower() == "a":
        if jenis_kopi:
            kopi = input("Masukkan Jenis kopi: ")
            ditemukan = False

            for i in range(len(jenis_kopi)):
                if jenis_kopi[i][0].lower() == kopi.lower():
                    input_stok = int(input("Masukkan jumlah stok (g): "))
                    jenis_kopi[i][1] += input_stok
                    print(f"Stok kopi {jenis_kopi[i][0]} berhasil ditambahkan. Totalnya: {jenis_kopi[i][1]}")
                    ditemukan = True
                    break

            if ditemukan == False:
                print(f"jenis kopi {kopi} tidak tersedia")
                
        else:
            print("Daftar jenis kopi kosong")
    elif pilih_tambah.lower() == "b":
        kopi = input("Masukkan Jenis kopi: ")
        input_stok = int(input("Masukkan jumlah stok (g)"))
        jenis_kopi.append([kopi, input_stok])
        print(f"{kopi} dengan stok {input_stok} berhasil ditambahkan")

def cek_ketersediaan():
    cek = input("Masukkan nama kopi yang ingin di cek: ")
    for nama, stok in jenis_kopi:
        if cek == nama:
            print(f"Stok kopi {nama} = {stok}")
        else:
            print("Jenis kopi tidak tersedia")

def hitung_kebutuhan():
    kopi = input("Masukkan jenis kopi: ")
    if jenis_kopi:
        for nama, stok in jenis_kopi:
            if kopi == nama:
                min_cup = int(input("Masukkan kebutuhan per cup: "))
                total_cup = stok // min_cup
                print(f"Total cup yang bisa di buat adalah: {total_cup}")
            else:
                print("Jenis kopi tidak tersedia")
    else:
        print("Daftar jenis kopi kosong")
                
while True:
    menu()
    pilih = input("Masukkan pilihan anda [1-5]: ")
    if pilih == "1":
        lihat_stok()
    elif pilih == "2":
        tambah_stok()
    elif pilih == "3":
        cek_ketersediaan()
    elif pilih == "4":
        hitung_kebutuhan()
    elif pilih == "5":
        break
    else:
        print("Input tidak valid")
```
