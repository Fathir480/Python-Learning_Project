Main Topic: List

-> List merupakan variabel bertipe list yang berisi kumpulan data

sintaks:
``` python
list_name = [data1, data2, data3, ...]
```

list method:
``` python
.append(data) # menambahkan data di akhir list
.insert(index, data) # menambahkan data berdasarkan index
.pop() # menghapus data di index terakhir
.remove(item) # menghapus item berdasarkan nama data
.clear() # mengosongkan list
.sort() # mengurutkan data
.reverse() # membaik urutan data
```

cara menampilkan list:
- menggunakan index
- for loops

````python
shopping_list = ['Milk', 'Eggs', 'Bread']
print(shopping_list[1])

shopping_list.append("Butter")
shopping_list.insert(1, "Juice")

print(shopping_list)

shopping_list.pop()
print(shopping_list)

for item in shopping_list:
    print(f"- {item}")
````

````python
for index, item in enumerate(shopping_list):
    print(f"{index + 1}. {item}")
````

````python
print(shopping_list)

shopping_list.sort()
print(shopping_list)

shopping_list.reverse()
print(shopping_list)

shopping_list.clear()
print(shopping_list)
````

````python
# Mini Project: Shopping List App

daftar_belanja = []

def menu_belanja():
    print("\n--- Daftar Menu Belanja ---")
    print("1. Lihat daftar belanja")
    print("2. Menambah barang belanja")
    print("3. Menghapus barang belanja")
    print("4. Membersihkan daftar belanja")
    print("5. Keluar")

while True:
    menu_belanja()
    pilihan = input("Masukkan pilihan Anda [1-5]: ")

    if pilihan == "1":
        print("\n--- Daftar Belanja ----")
        if not daftar_belanja:
            print("Tidak ada barang belanja")
        else:
            for index, item in enumerate(daftar_belanja):
                print(f"{index+1}. {item}")
    elif pilihan == "2":
        barang = input("Masukkan barang yang ingin ditambahkan: ")
        daftar_belanja.append(barang)
        print(f"{barang} berhasil ditambahkan")
    elif pilihan == "3":
        barang = input("Masukkan barang yang ingin dihapus: ")
        if barang in daftar_belanja:
            daftar_belanja.remove(barang)
            print(f"{barang}, berhasil dihapus")
        else:
            print(f"{barang}, tidak ada dalam daftar belanja")
    elif pilihan == "4":
        daftar_belanja.clear()
        print("Daftar belanja berhasil dibersihkan")
    elif pilihan == "5":
        print("Selamat Berbelanja!")
        break
    else:
        print("Pilihan tidak valid, silahkan masukkan pilihan yang tepat")
````
