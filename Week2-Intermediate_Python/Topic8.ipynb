Main Topic: Dictionaries

-> Sebuah struktur data yang menyimpan kumpulan data dalam bentuk pasangan key : value

sintaks:
``` python
my_dict = {
    "key1" : "value1",
    "key2" : "value2"
}
```

````python
contact = {
    "name" : "Budi",
    "phone" : "081234567890",
    "email" : "budi10@gmail.com"
}
print(contact.get("email")) # print email

contact["name"] = "Laura" # Ubah dict
print(contact)

contact["address"] = "Jl. Merpati No. 22b" # menambah dict
print(contact)

del contact["email"] # hapus dict
print(contact)

for key, value in contact.items():
    print(f"{key}: {value}")

if "email" in contact:
    print("Email found!")
else:
    print("Email not found!")
````

````python
# Mini project: Contact Book

contacts = {}

def show_menu():
    print("\n--- Contact Book Menu ---")
    print("1. Add Contact")
    print("2. View Contact")
    print("3. Search Contact")
    print("4. Edit Contact")
    print("5. Delete Contact")
    print("6. Exit")

def add_contact():
    name = input("Enter contact name: ")
    phone = input("Enter contact number:")
    email = input("Enter email: ")
    contacts[name] = {"phone":phone, "email":email}
    print(f"Contact {name} has been added successfully")

def view_contact():
    if not contacts:
        print("Contacts is Empty!")
    else:
        print("\n--- Contact List ---")
        for name, details in contacts.items():
            print(f"Name: {name}")
            print(f"Phone: {details['phone']}")
            print(f"Email: {details['email']}")

def search_contact():
    find_name = input("Enter contact name: ")
    if find_name in contacts:
        print(f"Phone: {contacts[find_name]['phone']}")
        print(f"Email: {contacts[find_name]['email']}")
    else:
        print(f"{find_name} not found!")

def edit_contact():
    edit_name = input("Enter contact name: ")
    if edit_name in contacts:
        phone = input("Enter new phonr number: ")
        email = input("Enter new email: ")
        contacts[edit_contact] = {"phone":phone, "email":email}
        print(f"Contact {edit_contact} has been updated successfully")
    else:
        print(f"{edit_name} not found!")

def delete_contact():
    delete_name = input("Enter contact name: ")
    if delete_name in contacts:
        del contacts[delete_name]
        print(f"Contact {delete_name} has been deleted successfully")
    else:
        print(f"{delete_name} not found!")

while True:
    show_menu()
    choice = input("Enter your choice [1-6]: ")
    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contact()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        edit_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
        print("\n--- Thanks you ---")
        break
````
