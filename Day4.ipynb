### Logical Operation

```python
a = 7
b = 12

if a > 5 and b < 15:
    print("Both condition are true")
else:
    print("At least one condition is False")
```

### Number Comparison Tool

```python
age = int(input("Enter your age:"))

print("---- Comparison Result ---")
if age <= 2:
    print("Your Kategori is Toddler")
elif age <= 18:
    print("Your Kategori is Teenager")
elif age <= 60:
    print("Your Kategori is Mature")
else:
    print("Your Kategori is Elderly")
```
