```python
# Using time.sleep for delay
import time

for i in range(5, 0, -1):
    print(i)
    time.sleep(1)
print("Happy Birth Day")
```

```python
# Countdown Timer

import time

start = int(input("Enter the number to start teh countdown time: "))
print("\n--- Countdown Begin ---")
while start > 0:
    print(start)
    time.sleep(1)
    start -= 1
print("Happy Birth Day")
```
